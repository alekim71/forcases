```python
from Bio.Seq import Seq
import IPython.display as display
import ipywidgets as widgets

# DNA sequence converter functions
def generate_complementary_sequence(dna_sequence):
    sequence = Seq(dna_sequence)
    complementary_sequence = sequence.complement()
    return complementary_sequence

def transcribe_to_mrna(dna_sequence):
    sequence = Seq(dna_sequence)
    mrna_sequence = sequence.transcribe()
    return mrna_sequence

def translate_to_amino_acids(mrna_sequence):
    sequence = Seq(mrna_sequence)
    amino_acid_sequence = sequence.translate()
    return amino_acid_sequence

# Create input widget
protein_input = widgets.Text(placeholder="Enter DNA sequence")
convert_button = widgets.Button(description="Convert")

# Output widgets
complementary_sequence_output = widgets.Output()
mrna_sequence_output = widgets.Output()
amino_acid_sequence_output = widgets.Output()

# Conversion function
def convert_sequence(button):
    protein_sequence = protein_input.value

    # Generate complementary sequence
    complementary_sequence = generate_complementary_sequence(protein_sequence)
    with complementary_sequence_output:
        display.display(complementary_sequence)

    # Transcribe to mRNA sequence
    mrna_sequence = transcribe_to_mrna(complementary_sequence)
    with mrna_sequence_output:
        display.display(mrna_sequence)

    # Translate to amino acid sequence
    amino_acid_sequence = translate_to_amino_acids(mrna_sequence)
    with amino_acid_sequence_output:
        display.display(amino_acid_sequence)

# Attach conversion function to button click event
convert_button.on_click(convert_sequence)

# Create layout
layout = widgets.VBox([
    widgets.HTML("<h1>DNA to AMINO ACID CONVERTER</h1>"),
    widgets.HTML("<p>Enter DNA sequence:</p>"),
    protein_input,
    convert_button,
    widgets.HTML("<br>"),
    widgets.HTML("<h3>Complementary Sequence</h3>"),
    complementary_sequence_output,
    widgets.HTML("<h3>mRNA Sequence</h3>"),
    mrna_sequence_output,
    widgets.HTML("<h3>Amino Acid Sequence</h3>"),
    amino_acid_sequence_output
])

# Display layout
display.display(layout)
```
